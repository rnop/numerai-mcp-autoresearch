```python
from pathlib import Path
from numerapi import NumerAPI

# If your notebook is inside the repo, this is enough:
ROOT = Path.cwd()

# If not, uncomment and set the repo root explicitly:
# ROOT = Path(r"C:\Users\nopro\OneDrive\Desktop\numerai-mcp-autoresearch")

DATA_VERSION = "v5.2"
DATA_DIR = ROOT / "data" / "numerai" / DATA_VERSION
DATA_DIR.mkdir(parents=True, exist_ok=True)

remote = f"{DATA_VERSION}/validation.parquet"
local = DATA_DIR / "validation.parquet"

print(f"Downloading {remote} -> {local}")
napi = NumerAPI()
napi.download_dataset(remote, str(local))

print("Done.")
print(local)
```
